```python
# percentage of male and female
gender = [55 , 45]
# counting of people at each age interval shown in age labels
age = [16 , 23, 347 , 428 , 334 , 398 , 923 , 1015 , 216 , 11]
age_labels = ['0-9','10-19','20-29','30-39','40-49','50-59','60-69','70-79','80-89','90-99']

plt.bar(age_labels, age)
plt.plot(age_labels, age, '--ro')
plt.xlabel('Age', fontweight='bold')
plt.ylabel('Quantity', fontweight='bold')

plt.figure(2)
plt.pie(gender, labels=['Male', 'Female'], explode=[0.05,0],autopct='%1.2f%%', shadow=True)
plt.title('Gender', fontweight='bold')
```

```python
from scipy.stats import norm

age_vector = []
for idx,age_count in enumerate(age):
    age_vector += [idx for x in range(age_count)]

vec = np.arange(min(age_vector), max(age_vector), 0.01)

plt.plot(vec, norm.pdf(vec, np.mean(age_vector), np.std(age_vector)))
plt.title('Age Distribution')
plt.ylabel('Probability Density')
```
